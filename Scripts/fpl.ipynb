```python
import fpl
import fbref
from datetime import datetime
import pandas as pd
import requests
import json
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
#url 
# Make sure to keep the trailing “/” in the url!
url = 'https://fantasy.premierleague.com/api/bootstrap-static/'
response = fpl.get(url)
```

```python
goalies,defenders,midfielders,attackers = fpl.team_stats(response)

midfielders_stats = fpl.midfielders_stats(response,midfielders)
attackers_stats = fpl.attackers_stats(response,attackers)
defenders_stats = fpl.defenders_stats(response,defenders)
goalies_stats = fpl.goalies_stats(response,goalies)
```

```python
overall_goalies,overall_defenders,overall_midfielders,overall_attackers = fpl.overall_player_stats(response)

overall_midfielders_stats = fpl.midfielders_stats(response,overall_midfielders)
overall_attackers_stats = fpl.attackers_stats(response,overall_attackers)
overall_defenders_stats = fpl.defenders_stats(response,overall_defenders)
overall_goalies_stats = fpl.goalies_stats(response,overall_goalies)
```

```python
overall_midfielders_stats = overall_midfielders_stats.assign(f = overall_midfielders_stats["total_points"]+overall_midfielders_stats["xPoints"]+overall_midfielders_stats["total_points_Home/Away"]+overall_midfielders_stats["xPoints_Home/Away"]).sort_values('f',ascending=False).drop('f', axis=1)
overall_midfielders_stats.head(20)
```

```python
overall_attackers_stats = overall_attackers_stats.assign(f = overall_attackers_stats["total_points"]+overall_attackers_stats["xPoints"]+overall_attackers_stats["total_points_Home/Away"]+overall_attackers_stats["xPoints_Home/Away"]).sort_values('f',ascending=False).drop('f', axis=1)
overall_attackers_stats.head(20)
```

```python
overall_defenders_stats = overall_defenders_stats.assign(f = overall_defenders_stats["total_points"]+overall_defenders_stats["xPoints"]+overall_defenders_stats["total_points_Home/Away"]+overall_defenders_stats["xPoints_Home/Away"]).sort_values('f',ascending=False).drop('f', axis=1)
overall_defenders_stats.head(20)
```
